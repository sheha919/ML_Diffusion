```python
# Import libraries for data manipulation
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_original =  pd.read_excel("D_cal.xlsx")
```

```python
# Define a list of column names to keep
keep_cols = ['index','Name','Density', 'individual-based Dxx', 'err Dxx', 'individual-based Dyy','err Dyy','individual-based Dx2d','err Dx2d','individual-based Dy2d','err Dy2d','individual-based D2d','err D2d']

# Use drop to select only the columns in the keep_cols list
df = df_original.drop(columns=[col for col in df_original.columns if col not in keep_cols])
```

```python
df = df.dropna(subset=['Name'])
```

```python
# Define a list of row indexes to drop
drop_rows = [20, 21]
# Use drop to remove the rows in the drop_rows list
df = df.drop(index=drop_rows)
```

```python
df = df.drop(index=range(0, 13))
df = df.drop(index=[41,42,43,44])
# Reset the index
df = df.reset_index(drop=True)
```

```python
water_12 = df[df['index'] == 'water']

# Plot the data points with error bars
#plt.errorbar(water_data['Density'], water_12['individual-based D2d'], yerr=water_data['err D2d'], fmt='--o',capsize=5)
plt.errorbar(water_12['Density'], water_12['individual-based Dx2d'], yerr=water_12['err Dx2d'], fmt='--o',capsize=5)
plt.errorbar(water_12['Density'], water_12['individual-based Dy2d'], yerr=water_12['err Dy2d'], fmt='--o',capsize=5)

# Add axis labels and a title
plt.xlabel('Density')
plt.ylabel('D')
plt.title('12.664')

# Show the plot
plt.savefig('C:/Users/Sachith/Desktop/work_oh/separation_12_density_with_error.jpg', dpi=400)
plt.show()
```
