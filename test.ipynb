```python
kmean = np.mean(kenergy[st:end])
kstd = np.std(kenergy[st:end])
print('mean kinetic energy   =',kmean)
print('std of kinetic energy =',kstd)
```

```python
pmean = np.mean(penergy[st:end])
pstd = np.std(penergy[st:end])
print('mean potential energy   =',pmean)
print('std of potential energy =',pstd)
```

```python
tmean = np.mean(tenergy[st:end])
tstd = np.std(tenergy[st:end])
print('mean total energy   =',tmean)
print('std of total energy =',tstd)
```

```python
nmovingatoms=32
print('start    at                                   = ',st)
print('ends at                                       = ',end)
print('number of moving atoms                        =',nmovingatoms) #'{0: >#016.8f}'.format()
print('mean kinetic energy                           =','{0: >#026.18f}'.format(kmean))
print('std of kinetic energy                         =','{0: >#026.18f}'.format(kstd))
print('mean potential energy                         =','{0: >#026.18f}'.format(pmean))
print('std of potential energy                       =','{0: >#026.18f}'.format(pstd))
print('mean total energy                             =','{0: >#026.18f}'.format(tmean))
print('std of total energy                           =','{0: >#026.18f}'.format(tstd))
print('------------------------------------------------------------------------')
print('temperature from mean kinetic energy          = ',kmean*2*315777/(nmovingatoms*3),'K')
print('thermal fluctuations of the totalenergy       = ',1*100/np.sqrt(3*nmovingatoms),'%')
print('total energy fluctuations compared to 3N*KT   = ',tstd*100/(kmean*2),'%')
print('kinetic energy fluctuations compared to 3N*KT = ',kstd*100/(kmean*2),'%')
```

```python

def moving_average(data, window_size):
    return np.array([np.mean(data[i:i+window_size]) for i in range(len(data) - window_size + 1)])

# Assuming tenergy is your data array and you want a moving average over 10 data points
window_size = 20
moving_avg = moving_average(tenergy, window_size)
```
