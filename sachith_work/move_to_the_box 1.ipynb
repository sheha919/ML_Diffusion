```python
atom= ['Non']*(natoms)
x_coordinate =np.empty(natoms)
y_coordinate = np.empty(natoms)
z_coordinate = np.empty(natoms)
x_vel =np.empty(natoms)
y_vel =np.empty(natoms)
z_vel =np.empty(natoms)
charge =np.empty(natoms)
```

```python
ii=0
f=open(path2 +'/finalstep.xyz','r')
lines=f.readlines()
for line in lines[:]:
    data=line.split()
    if len(data)>3:
        atom[ii]=data[0]
        x_coordinate[ii] = float(data[1])
        y_coordinate[ii] = float(data[2])
        z_coordinate[ii] = float(data[3])
        charge[ii] =  float(data[4])
        x_vel[ii] = float(data[5])
        y_vel[ii] = float(data[6])
        z_vel[ii] = float(data[7])
        ii = ii+1
```

#### put the atom index -1 and the direction of the atom need to be moved

```python
mvinxp = []
mvinxn = [] #moved in negative x (atom was out of the box in positive x)
mvinyp = []
mvinyn = []
mvinzp = []
mvinzn = []
```

```python
ijk = 1
while ijk < 2:
    print(ijk)
    ijk = ijk+1
    jj = 0
    while jj < natoms :
        if x_coordinate[jj]>(xbox/2):
            x_coordinate[jj]=x_coordinate[jj]-xbox
            mvinxn.append(jj+1)
            ijk = 1
        if x_coordinate[jj]<(-xbox/2):
            x_coordinate[jj]=x_coordinate[jj]+xbox
            mvinxp.append(jj+1)
            ijk = 1
        
        if y_coordinate[jj]>(ybox/2):
            y_coordinate[jj]=y_coordinate[jj]-ybox
            mvinyn.append(jj+1)
            ijk = 1
        if y_coordinate[jj]<(-ybox/2):
            y_coordinate[jj]=y_coordinate[jj]+ybox
            mvinyp.append(jj+1)
            ijk = 1
            
        if z_coordinate[jj]>(zbox/2):
            z_coordinate[jj]=z_coordinate[jj]-zbox
            mvinzn.append(jj+1)
            ijk = 1
        if z_coordinate[jj]<(-zbox/2):
            z_coordinate[jj]=z_coordinate[jj]+zbox
            mvinzp.append(jj+1)
            ijk = 1
        
        jj=jj+1
```

```python
f=open(path2 +'/moved_atoms.dat','w')
print('moved to negative x',mvinxn,file =f)
print('moved to positive x',mvinxp,file =f)
print('moved to negative y',mvinyn,file =f)
print('moved to positive y',mvinyp,file =f)
f.close()
```

```python
kk=0
f=open(path2 +'/finalstep_InTheBox.xyz','w')
print(natoms,file =f)
f.write("\n")
while kk < natoms :
    print('{:3}'.format(atom[kk]),'{0: >#016.8f}'.format(x_coordinate[kk]),'{0: >#016.8f}'.format(y_coordinate[kk]),'{0: >#016.8f}'.format(z_coordinate[kk]),'{0: >#016.8f}'.format(charge[kk]),'{0: >#016.8f}'.format(x_vel[kk]),'{0: >#016.8f}'.format(y_vel[kk]),'{0: >#016.8f}'.format(z_vel[kk]),file =f )
    kk=kk+1
f.close()
```
